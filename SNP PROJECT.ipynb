```python
##DEFINE SEQUENCE TO COMPARE TWO DNA SEQUENCES. OUR ORIGINAL DATA SETS 
##EACH HAD OVER 6 MILLION CHARACTERS --> COULD NOT RUN IN NOTEBOOK DUE 
##TO DATA SET SIZE & JUPYTER NB'S PROCESSING POWER --> SHORTENED THE 
##DATA SETS AND ADDED 2 MUTATIONS TO SHOW AN EXAMPLE OF HOW THIS 
##FUNCTION COULD BE IMPLEMENTED WITH REAL DNA SEQUENCES

##INPUT = SEQUENCE OF DNA DATA
##OUTPUT = SNP NUCLEOTIDE LOCATION

def sequence_comp(R1, R2):
    len1 = len(R1)
    len2 = len(R2)
    trueorfalse = []
    onlymismatch = []
    for p in range(0, min(len1,len2)):
        if R1[p] != R2[p]:
            add = "FALSE", p 
            trueorfalse.append(add)
            onlymismatch.append(add)
        else:
            trueorfalse.append("T")
    print("The SNP location is:", onlymismatch)
    print(trueorfalse)
    

sequence_comp(R1,R2)

```
