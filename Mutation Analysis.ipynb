```python
mutations = ('PTR', 'PAD', 'CHP', 'HP', 'PBPA', 'TRV', 'TT', 'TF', 'ETC')
R1 = [7, 3, 8, 26, 4, 1, 5, 3, 13]
R2 = [6, 3, 8, 26, 4, 2, 5, 4, 13]
BAA47 = [7, 3, 8, 22, 4, 1, 5, 3, 23]

data = pd.DataFrame({'R1': R1,'R2': R2,'BAA47': BAA47,}, index = mutations)

bar_graph = data.plot.bar(rot=0)

plt.title('Mutations in Pseudomonas aeruginosa', fontsize = 14)
plt.ylabel('Total', fontsize = 14)
plt.xlabel('Mutation', fontsize = 14)

plt.show()
```
